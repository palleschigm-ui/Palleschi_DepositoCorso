Quando usi df.describe() su questo dataset pandas riconosce tutte le colonne come oggetti (stringhe), quindi produce automaticamente le statistiche descrittive per variabili categoriche, non per variabili numeriche.

Per questo vedi:

count → numero di valori non nulli

unique → numero di valori diversi

top → valore più frequente

freq → frequenza del valore più comune

Queste sono le statistiche standard per variabili di tipo object.
Tutte e 11 le variabili sono lette come object, anche quelle che dovrebbero essere numeriche (HorsePower, Speed, Prices, Torque ecc.).

Quindi df.describe() interpreta tutto come testo e ti mostra statistiche per categorie.
Perché succede?

Perché nel tuo CSV:

i numeri probabilmente contengono unità di misura (es. “355 hp”, “250 km/h”, “400 Nm”)

o simboli come $, km/h, cc, hp, Nm, sec

il separatore dei decimali non è standard

oppure i valori sono misti (testo + numeri)

Quindi pandas NON può convertirli automaticamente in numerici.
